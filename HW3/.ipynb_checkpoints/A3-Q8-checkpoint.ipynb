```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st 
import math
```

```python
#code for Q7a
def plot_a(x, plot):
    
    n = len(x)
    x.sort()
    x_a = []
    y_a = []
    y_curr = 0        
    x_a.append(0)
    y_a.append(0)

    for i in x:
        y_curr += 1/n
        y_a.append(y_curr)
        x_a.append(i)
        
    if plot:
#         plt.figure("Empirical CDF")
        plt.grid(True)
        plt.step(x_a, y_a, where="post", label = "F_Hat")
        plt.xlabel("X")
        plt.ylabel("Pr[X<alpha]")

    x_a = x_a[1:]
    y_a = y_a[1:]
    
    return (x_a, y_a)
```

### **PART (a): Modify the program from 7 (a) to now also add 95% Normal-based CI lines for $\hat{F_n}$, using the resultfrom 6 (b), given a list of samples as input. Draw a plot showing $\hat{F_n}$ and the CI lines for the q8.dat data file (799 samples) on the class website. Use x-limits of 0 to 2, and y-limits of 0 to 1.**

```python
#Q8a
def normal_ci(x_a,y_a,z,show):
    fHat = np.asarray(y_a)        
    n = fHat.shape[0]
    fHat = np.around(fHat, 4)
    subTerm = np.subtract(1,fHat)
    multTerm = np.around(np.multiply(fHat,subTerm),4)
    divTerm = np.divide(multTerm, float(n))    
    sqrtTerm = np.sqrt(divTerm)    
    intv = np.multiply(z,sqrtTerm)
    #f_hat - z(alpha/2)*sqrt(f_hat*(1-f_hat)/n)
    y_a1 = np.subtract(fHat, intv)
    #f_hat + z(alpha/2)*sqrt(f_hat*(1-f_hat)/n)
    y_a2 = np.add(fHat, intv)
    
    # Plotting
    
#     plt.figure(figsize=(20,10))
    plt.step(x_a, y_a1, where="post", label = "normal lower CI")
    plt.step(x_a, y_a2, where="post", label ="normal upper CI")
    plt.grid(True)
    plt.title("Empirical CDF and CI Lines")
    plt.xlim([0,2])
    if show:
        plt.legend(bbox_to_anchor=(1, 0), loc=4, borderaxespad=0.)
        plt.show()
```

### **PART (b): Modify the program from 8 (a) above to also add 95% DKW-based CI lines for $\hat{F_n}$. Draw a plot showing $\hat{F_n}$ and both CI lines (Normal and DKW) for the q8.dat data. Which CI is tighter?**

```python
#Q8b
def dkw_ci(x_a, y_a, alpha): 
    # epsilon = sqrt(1/2n*ln(2/alpha))
    fHat = np.asarray(y_a)        
    n = fHat.shape[0]
    fHat = np.around(fHat, 4)   
    epsilon = math.sqrt(math.log(2/float(alpha))/float(2*n))    
    # f_hat - epsilon
    y_d1 = np.subtract(fHat, epsilon)
    # f_hat + epsilon
    y_d2 = np.add(fHat, epsilon)
    
    # Plotting
#     plt.figure(figsize=(20,10))
    plt.step(x_a, y_d1, where="post", label = "DKW lower CI")
    plt.step(x_a, y_d2, where="post", label = "DKW upper CI")
    plt.grid(True)
    plt.xlim([0,2])
    plt.title("Empirical CDF and CI Lines")
    plt.legend(bbox_to_anchor=(1, 0), loc=4, borderaxespad=0.)
    plt.show()
```
